```python
import pandas as pd
```

```python
sanakirja = {
    "id":[1,2,3,4,5],
    "ika":[20,34,76,58,44],
    "sukupuoli":['N','M','M','N','M']
}
df = pd.DataFrame(sanakirja)
df
```

```python
df['pituus'] = [158,164,170,165,161]
df
```

```python
df2 = pd.DataFrame(
    {
        'kengennumero':[40,45,42,41,39],
        'paino' : [60,80,75,69,12]
    }
)
df2
```

```python
df3 = pd.concat([df,df2],axis=1)
df3
```

```python
df3.loc[1] = [2,32,'N',173,38,70]
df3
```

```python
df3.index
```

```python
df3.loc[len(df3.index)] = [6,15,'N',180,39,68]
df3
```

```python
df_series = pd.Series([7,64,'M',175,43,75],index=df3.columns)
df_series
```

```python
apu_df = pd.DataFrame([df_series])
apu_df
```

```python
df4 = pd.concat([df3,apu_df],ignore_index=True)
df4
```

Data editing

```python
df6.loc[(df6['sukupuoli']=='M'),'paino'] = df6.loc[(df6['sukupuoli']=='M'),'paino'] + 23
df6
```

```python
df8 = df6.drop(['ika','sukupuoli'],axis=1)
df8
```

```python
indeksi = df6[df6['id']==6].index
df9 = df6.drop(indeksi)
df9
```

```python
df10 = df9.dropna()
df10
```

```python
df10 = df10.drop_duplicates()
df10
```
