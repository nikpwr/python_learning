```python
array8 = np.random.normal(size=10, loc=0, scale=1)
array8
```

```python
print(f"Last value is {array8[-1]}")
```

```python
print('pienin arvo', np.min(array8))
print('suurin arvo', np.max(array8))
print('keskiarvo', np.mean(array8))
print('määrä', len(array8))
print('keskihajonta', np.std(array8))
print('otosvarianssi', np.var(array8,ddof=1))
```
